```python
import numpy as np
import gensim.downloader as api
```

```python
model = api.load("glove-wiki-gigaword-100")

```

```python
def cos_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

```

```python
def assoc(w, A, B):
    return np.mean([cos_sim(model[w], model[a]) for a in A]) - \
           np.mean([cos_sim(model[w], model[b]) for b in B])

```

```python
def weat(X, Y, A, B):
    return sum([assoc(x, A, B) for x in X]) - \
           sum([assoc(y, A, B) for y in Y])

```

```python
def effect(X, Y, A, B):
    X_assoc = [assoc(x, A, B) for x in X]
    Y_assoc = [assoc(y, A, B) for y in Y]
    return (np.mean(X_assoc) - np.mean(Y_assoc)) / np.std(X_assoc + Y_assoc)

male = ["man", "he", "him"]
female = ["woman", "she", "her"]

career = ["job", "career", "office"]
family = ["home", "family", "children"]

score = weat(male, female, career, family)
effect_size = effect(male, female, career, family)

print("WEAT Score:", score)
print("Effect Size:", effect_size)
```
